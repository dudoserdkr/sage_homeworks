## Завдання 1

Розгляньте групу матриць за множенням  

$$G = \left\langle \left(\begin{array}{rrr}
1 & 0 & 1 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{array}\right),  \left(\begin{array}{rrr}
1 & 0 & 0 \\
0 & 1 & 1 \\
0 & 0 & 1
\end{array}\right)\right\rangle$$

Дайте відповідь на наступні питання: 

1. _Чи є група G скінченною?_
2. _Чи є група G комутативною?_
3. _Який вигляд мають елементи групи G?_
4. _Які порядки мають елементи групи G?_ 
5. _Чому ізоморфна група G?_


__Відповіді__: 
1. Ні 
2. Так
3. $\forall g \in G, \ \forall k, m \in \mathbb{N} \cup \{0 \}: \ g = M_{1}^{k} \cdot M_{2}^{m}, \ g = \begin{pmatrix} 1 & 0 & k \\ 0 & 1 & m \\ 0 & 0 & 1 \end{pmatrix}$ 
4. Усі мають нескінченний порядок, окрім нейтрального
5. $G \simeq \mathbb{Z} \oplus \mathbb{Z}$ 

```python
M1 = Matrix([[1, 0, 1], [0, 1, 0], [0, 0, 1]])
M2 = Matrix([[1, 0, 0], [0, 1, 1], [0, 0, 1]])
G = MatrixGroup([M1, M2])

#1
print(G.cardinality())
#2
print(G.is_abelian())

#3
var("k", "m")
print(M1^k * M2^m)
print()
print((M1 * M2)^k)
print()
print(M1^k)
print()
print(M2^k)

```
