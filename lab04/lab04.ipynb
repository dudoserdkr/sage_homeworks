## Завдання 1. 

__Знайдіть усі незвідні системи твірних групи $S_4$.__

```python
# Для S4 система твірних буде максимум з 2 елементів
from itertools import combinations

S4 = SymmetricGroup(4)
all_elements = S4.list()
generating_systems = []

for element_pair in combinations(all_elements, 2):
    if S4.subgroup(element_pair) == S4:
        generating_systems.append(element_pair)

for gen_sys in generating_systems:
    print(gen_sys)
```

## Завдання 2. 

__Знайдіть усі підгрупи груп $S_4, A_4, S_5, A_5$.__

```python
class SubgroupPrinter:
    def __init__(self, n, group_objects):
        self.n = n
        self.group_objects = group_objects

    @staticmethod
    def print_subgroup(group):
        print(f"----------{group}----------")
        for subgroup in group.subgroups():
            print(subgroup)

    def slava_ukraini(self): # Я не знаю як назвати метод
        for obj in self.group_objects:
            for i in self.n:
                self.print_subgroup(obj(i))

n = [4, 5]
group_objects = [SymmetricGroup, AlternatingGroup]
SP = SubgroupPrinter(n, group_objects)

SP.slava_ukraini()
```

```python
from random import randint

def is_even_permutation(s):
    cycles = s.cycle_tuples()
    transpositions = sum(len(c) - 1 for c in cycles)
    return transpositions % 2 == 0

n = int(input())
An = AlternatingGroup(n)
Sn = SymmetricGroup(n)
s1, s2 = Sn.list()[randint(0, factorial(n) - 1)], Sn.list()[randint(0, factorial(n) - 1)]

for s in [s1, s2]:
    print(f"is {s} even: {is_even_permutation(s)}")
    
subgroup = Sn.subgroup([s1, s2])
print(f"|<{s1}, {s2}>| == {subgroup.order()}")
print(f"is <{s1}, {s2}> == Sn? - {subgroup == Sn}")
print(f"is <{s1}, {s2}> == An? - {subgroup == An}")

```
