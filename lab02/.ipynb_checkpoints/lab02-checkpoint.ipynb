## Завдання 2. 
__Для кожної з визначених дій перевірити чи буде дія асоціативною. Якщо дія не асоціативна, то вивести трійку, для якої ця умова порушується.__

```python
from itertools import permutations

class ActionGenerator:
    def __init__(self, needed_set):
        self.set = needed_set

    def _all_actions(self):
        M = FiniteSetMaps(product(self.set, repeat=2), self.set)
        for f in M:
            yield f
            
    def actions(self):
        number_of_action = 0
        for bin_action in all_actions(needed_set):
            number_of_action += 1
            
            def _q(x, y):
                return bin_action((x, y))
                
            OT = OperationTable(needed_set, _q, names="elements")
            yield (OT, number_of_action)    

with open("t02_output.txt", "w", encoding="utf-8") as file:
    sets = [[1, 2], [1, 2, 3]]
    for n in range(2):
        
        needed_set = sets[n]
        file.write(f"------------------Actions with n={n + 2}------------------\n")
        AG = ActionGenerator(needed_set)
        
        for OT, number_of_action in AG.actions():
            is_OT_associative = True
            all_permutations = permutations(needed_set, int(3))
            
            wrong_p = []
            for p in all_permutations:
                if OT[ OT[ p[0], p[1] ], p[2] ] != OT[ p[0], OT[ p[1], p[2] ] ]:
                    is_OT_associative = False
                    wrong_p = [int(d) for d in p]

                    break
                    
            file.write(f"Number of action: {number_of_action}\n")
            file.write(str(OT))
            s = ""
            if is_OT_associative:
                s = "is associative"
            else:
                s = "is not associative"
                
            file.write(f"This operation " + s + "\n")
            file.write(f"{wrong_p}\n\n")
                
```

## Завдання 3. 

__Перевірити, які з визначених множин з асоціативними бінарними діями будуть ізоморфними між собою і знайти кількість класів попарно ізоморфних між собою множин.__

```python
from itertools import permutations, product

needed_set = [1, 2, 3]

def all_actions(space):
    keys = list(product(space, repeat=2))
    for values in product(space, repeat=len(keys)):
        mapping = dict(zip(keys, values))
        yield mapping

def is_associative(space, action):
    for x in space:
        for y in space:
            for z in space:
                left = action[(x, action[(y, z)])]
                right = action[(action[(x, y)], z)]
                if left != right:
                    return False, [x, y, z]
    return True, None

associative_actions = []

for f in all_actions(needed_set): 
    result, counter = is_associative(needed_set, f)
    if result: 
        associative_actions.append(f)

with open('t03_output.txt', 'w') as file2:
    file2.write(f'К-ть асоціативних дій для {len(needed_set)}-елементної множини: {len(associative_actions)}\n')

def is_isomorphic(space1, action1, space2, action2):
    space1 = list(space1)
    space2 = list(space2)
    for perm in permutations(space2):
        mapping = {space1[i]: perm[i] for i in range(len(space1))}
        is_iso = True
        for a in space1:
            for b in space1:
                f1_result = action1[(a, b)]
                mapped_f1 = mapping.get(f1_result, None)
                f2_result = action2[(mapping[a], mapping[b])]
                if mapped_f1 != f2_result:
                    is_iso = False
                    break 
            if not is_iso:
                break 
        if is_iso:
            return True
    return False

isomorph_classes = []

for f1 in associative_actions: 
    for iso_class in isomorph_classes: 
        f2 = iso_class[0]
        if is_isomorphic(needed_set, f1, needed_set, f2): 
            iso_class[1] += 1
            break 
    else:    
        isomorph_classes.append([f1, 1])

with open('t03_output.txt', 'w') as file3:
    print('К-ть класів ізоморфності:', len(isomorph_classes), file = file3)
    print('К-ть дій в кожному класі:', file = file3)
    for iso_class in isomorph_classes: 
        print(*iso_class, file = file3)
```
